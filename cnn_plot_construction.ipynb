```python
#Plot Construction
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
# Vanilla CNN 8 Classes
with open("Vanilla CNN/8 classes/vanilla_cnn_8_training_acc", 'rb') as f:
    trainingAcc = pickle.load(f)
with open("Vanilla CNN/8 classes/vanilla_cnn_8_training_loss", 'rb') as f:
    trainingLoss = pickle.load(f)
with open("Vanilla CNN/8 classes/vanilla_cnn_8_val_acc", 'rb') as f:
    testingAcc = pickle.load(f)
with open("Vanilla CNN/8 classes/vanilla_cnn_8_val_loss", 'rb') as f:
    testingLoss = pickle.load(f)
```

```python
x = np.linspace(1,10,len(trainingAcc))

#mean training/testing acc during last epoch - 74 recordings per epoch for 8 classes
train_acc_avg = [sum(trainingAcc[-74:])/len(trainingAcc[-74:])]*len(trainingAcc)
test_acc_avg = [sum(testingAcc[-74:])/len(testingAcc[-74:])]*len(testingAcc)


plt.plot(x,trainingAcc, '-b', label = 'training', zorder=5)
plt.plot(x,testingAcc, '-r', label = 'validation', zorder = 0)
plt.plot(x,train_acc_avg, '--b', label = 'avg. training (last epoch) = %.2f' %train_acc_avg[0])
plt.plot(x,test_acc_avg, '--r', label = 'avg. validation (last epoch) = %.2f' %test_acc_avg[0])
plt.ylabel("accuracy")
plt.xlabel("epochs")
plt.title("Vanilla CNN Accuracy - 8 Classes")
plt.legend(loc='center right')
plt.grid()
plt.savefig('Vanilla CNN/8 classes/vanillaCNN_8class_acc.png')
plt.show()


#mean training/testing loss during last epoch - 74 recordings per epoch for 8 classes
train_loss_avg = [sum(trainingLoss[-74:])/len(trainingLoss[-74:])]*len(trainingLoss)
test_loss_avg = [sum(testingLoss[-74:])/len(testingLoss[-74:])]*len(testingLoss)


x = np.linspace(1,10,len(trainingLoss))

plt.plot(x,trainingLoss, '-b', label = 'training',zorder=5)
plt.plot(x,testingLoss, '-r', label = 'validation',zorder=0)
plt.plot(x,train_loss_avg, '--b', label = 'avg. training (last epoch) = %.3f' %train_loss_avg[0])
plt.plot(x,test_loss_avg, '--r', label = 'avg. validation (last epoch) = %.3f' %test_loss_avg[0])
plt.ylabel("loss")
plt.xlabel("epochs")
plt.title("Vanilla CNN Loss - 8 Classes")
plt.legend( loc='upper right')
plt.grid()
plt.savefig('Vanilla CNN/8 classes/vanillaCNN_8class_loss.png')
plt.show()
```
